Part 1 — if/else

The finance hook

Every trading rule you've ever used is an if/else statement in disguise.

"If price breaks above the high of the previous session, go long. Otherwise, stay flat."

Python just lets you write that logic formally.

Three things to notice before you run it:

The condition after if has no brackets required

The line ends with a colon :

The code inside is indented — 4 spaces or one Tab. Python uses indentation to know what belongs inside the if block

```python
pnl = 500

if pnl > 0:
  print("Winning trade")
else:
  print("Losing trade")
```

elif — the middle case

elif means "else if" — check this condition only if the first one was false.

The structure is always:

if — first check

elif — second check (as many as you need)

else — everything that didn't match above

Run it with pnl = 0, then try -500 and +750. Tell me all three outputs.

```python
pnl = 0

if pnl > 0:
  print("Winning trade")
elif pnl == 0:
  print("Breakeven")
else:
  print("Losing trade")

```

```python
trade1 = {
  "Direction": "Long",
  "Entry": 5842.25,
  "Exit": 5851.50,
  "Contracts": 3
}

pnl = (trade1["Exit"] - trade1["Entry"]) * trade1["Contracts"] * 50

if pnl > 0:
  trade1["Result"] = "Winner"
elif pnl == 0 :
  trade1["Result"] = "Breakeven"
else:
  trade1["Result"] = "Loser"

trade1["Pnl"] = pnl
print(trade1)
```

 The dictionary labelled itself based on the numbers — you didn't have to type "Winner" manually.

This is exactly what happens in your ML notebooks. In P10, the logistic regression notebook labels each day as "up" or "down" based on price movement using this same logic. You just wrote the manual version of that.

What if the journal had 500 trades? You'd need 500 print lines.

A for loop says: do this once for every item in the list, automatically.

```python
closes = [5260.75, 5271.00, 5268.50, 5280.25, 5285.50]

for price in closes:
  print(price)
```

price is a temporary variable that takes the value of each item one at a time. The loop runs 5 times — once per item.

```python
for price in closes:
  if price > 5270:
    print(price, "- above 5270 ")
  else:
    print(price, "- below 5270 ")
```

Refresher

```python
trade1 = {
    "Direction": "Long",
    "Entry": 5842.25,
    "Exit": 5851.50,
    "Contracts": 3
}

trade2 = {
    "Direction": "Short",
    "Entry": 5856.00,
    "Exit": 5847.25,
    "Contracts": 2
}

trade3 = {
    "Direction": "Long",
    "Entry": 5848.75,
    "Exit": 5838.00,
    "Contracts": 4
}

pnl = (trade1["Exit"] - trade1["Entry"]) * trade1["Contracts"] * 50
trade1["Pnl"] = pnl
pnl = (trade2["Entry"] - trade2["Exit"]) * trade2["Contracts"] * 50
trade2["Pnl"] = pnl
pnl = (trade3["Exit"] - trade3["Entry"]) * trade3["Contracts"] * 50
trade3["Pnl"] = pnl

journal = [trade1, trade2, trade3]

for trade in journal:
  if trade["Pnl"] > 0:
    print (trade["Direction"], "|", trade["Pnl"], "is a Winner")
  elif trade["Pnl"] == 0:
    print (trade["Direction"], "|", trade["Pnl"], "is a Breakeven")
  else:
    print (trade["Direction"], "|", trade["Pnl"], "is a Loser")
```

trade acts as a temp placeholder for us to loop the journal
inside the journal we have trade1, trade2 and trade3

so that "trade" will be replace by trade1, trade2 and trade3 while looping

therefore we are basically printing
print (trade1["Direction"], "|", trade1["Pnl"])

The variable name trade is just a label you choose — you could call it anything. for t in journal, for row in journal, for banana in journal — all work identically. Python doesn't care what you name it. What matters is that it holds one item from the list on each pass.
