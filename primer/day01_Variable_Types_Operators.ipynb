The finance hook first
As a futures trader, every time you look at your screen you see numbers. ES price: 5,285.50. Your P&L: -$312.50. Position size: 2 contracts.

In Python, when you want to work with a number — calculate something, store it, compare it — you give it a name. That name is called a **variable**

```python
es_price = 5285.50
pnl = -312.50
contracts = 2
```

That's it. That's a variable. A name that holds a value.

```python
print(es_price)
print(pnl)
print(contracts)
```

What just happened — line by line:

| Line               | What it does                                                        |
| ------------------ | ------------------------------------------------------------------- |
| es_price = 5285.50 | Creates a box called es_price and puts the number 5285.50 inside it |
| pnl = -312.50      | Creates a box called pnl and puts -312.50 inside it                 |
| contracts = 2      | Creates a box called contracts and puts 2 inside it                 |
| print(es_price)    | Opens the box and shows you what's inside                           |

```python
total_value = es_price * contracts
print(total_value)
```

You just used two variables to calculate a third one. Python remembered what es_price and contracts were from the cell above.

```python
#print(ES_price)
```

You'll get an error. Python is case-sensitive. es_price and ES_price are two completely different names. This is one of the most common beginner bugs — and now you know it exists.

Data Types

Here's the thing about variables. Python doesn't just store numbers. It stores different kinds of things. And the kind matters — because you can't do the same operations on all of them.

The finance hook
Think about your trading screen again. You see:

5285.50 — a price (a number with decimals)

2 — number of contracts (a whole number)

"ES" — the instrument name (text)

True / False — is the trade a winner? (yes or no)

Python has a specific type for each of these.

```python
price = 5285.50        # float — numbers with decimals
contracts = 2          # int — whole numbers only
instrument = "ES"      # str — text, always in quotes
is_winner = True       # bool — only True or False

print(type(price))
print(type(contracts))
print(type(instrument))
print(type(is_winner))
```

type() is Python telling you what kind of thing is inside the box.

***Why this matters in your ML notebooks***

Open P9 in your head for a second. When the notebook loads price data, every column has a type. If a price column accidentally gets stored as str instead of float, your model breaks. The error won't say "wrong type" — it'll say something cryptic. Knowing types is how you debug that.

```python
price = 5285.50
contracts = 2

total = price * contracts
print(total)
print(type(total))
```

```python
price = "5285.50"
contracts = 2

total = price * contracts
print(total)
```

```python
price = "5285.50"
price_fixed = float(price)   # convert str → float
contracts = 2

total = price_fixed * contracts
print(total)
print(type(price_fixed))
```

```python
price = 5285.50
prev_close = 5270.00

#The 6 operators you'll use constantly
difference = price - prev_close
move_pct = (difference/prev_close) * 100
squared = difference ** 2
contracts = 10
remainder = contracts % 3 #modulo - remainder after division

print("Point move:   ", difference)
print("Percent move: ", move_pct)
print("Squared:      ", squared)
print("Remainder:    ", remainder)
```

Session 1 wrap-up

Variables — named boxes that store values. Case-sensitive. = means store, not equals.

Operators — +, -, *, /, **, %. You already think in these as a trader — Python just formalises them.

Quotes = text. No quotes = number. Mixing them doesn't always crash — sometimes it does something weird and silent. That's the dangerous bug.
