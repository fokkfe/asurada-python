Indexing — accessing one item
You have 5 closes. How do you grab just one?

Python numbers every item in a list starting from 0, not 1. This trips up almost every beginner at least once.

closes = [5260.75, 5271.00, 5268.50, 5280.25, 5285.50]
index:        0        1        2        3        4

Think of it as: index = how many steps from the beginning.

```python
closes= [5260.75, 5271.00, 5268.50, 5280.25, 5285.50]

print(closes[0])
print(closes[4])
print(closes[-1])
print(closes[2])
```

 notice closes[4] and closes[-1] both give you the same number — two different ways to reach the last item. You'll see -1 used constantly in ML code because the data length is often unknown, but the last item is always -1.

*** Slicing — grabbing a range ***

Now what if you want the last 3 closes — like looking at the most recent 3 bars?

```python
closes= [5260.75, 5271.00, 5268.50, 5280.25, 5285.50]

print(closes[2:5]) #NOT INCLUDING 5 -> 2, 3, 4
print(closes[-3:]) # the last three items -> -3, -2, -1
print(closes[:3]) #first 3 items -> 0, 1 , 2
```

The syntax is list[start:stop]. The stop index is not included. So [2:5] gives you index 2, 3, 4 — three items

Notice closes[2:5] and closes[-3:] give you the exact same result — just two different ways to say "last 3 items." In ML code you'll see [-3:] more often because you don't always know the length of the list in advance.

*** Modifying a list ***

Lists are not fixed. You can add, change, and remove items.

```python
closes = [5260.75, 5271.00, 5268.50, 5280.25, 5285.50]

# Add a new close at the end
closes.append(5290.00)
print(closes)

#change the first close
closes[0] = 5261.00
print(closes)

#remove the last item
closes.pop()
print(closes)
```

*** Mixed types ***

```python
trade = ["ES", 5285.50, 2, True, "long"]

print(trade[0])
print(trade[1])
print(type(trade[0]))
print(type(trade[1]))
```

A list can hold any mix of types — strings, floats, ints, bools, all together

But here's the thing. In ML and data work, you almost never want mixed-type lists. Your price list should be all floats. Your label list should be all ints. Mixed types cause the same kind of silent bugs as the string/number issue from Session 1.

Pure lists → clean data → models that don't break.
