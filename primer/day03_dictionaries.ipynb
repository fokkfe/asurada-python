*** Dictionaries ***

The finance hook
A list stores values in order — but it has no labels. If someone gives you:

[5285.50, 2, -312.50, "long"]

You have to remember that index 0 is price, index 1 is contracts, index 2 is P&L. That's fragile.

A dictionary solves this. Instead of accessing by position, you access by name.

Think of it like your trading blotter — each trade has named fields: instrument, price, size, direction.

```python
trade = {
    "instrument": "ES",
    "entry_price": 5285.50,
    "contracts": 2,
    "pnl": -312.50,
    "direction": "long"
}

print(trade)
print(type(trade))
```

The dictionary itself is type dict — but inside it, each value keeps its own type. "ES" is still a str, 5285.50 is still a float, 2 is still an int. The dictionary is just the container that organises them with names.

dictionary vs list

With a list you used closes[0]. With a dictionary you use the key name instead of a number.

```python
print(trade["instrument"])
print(trade["entry_price"])
print(trade["pnl"])
```

```python
#change an existing value

trade["entry_price"] = 5290.00
print(trade["entry_price"])

#add a brand new key
trade["stop_loss"] = 5275.00
print(trade)

#remove a key
del trade["direction"]
print(trade)
```

looping through a dictionary

```python
trade = {                            #wiggle bracket
    "instrument": "ES",              #atttention to colon and comma #str need to be bracketed
    "entry_price": 5290.00,
    "contracts": 2,
    "pnl": -312.50,
    "stop_loss": 5275.00              #ends with no comma
}

for key, value in trade.items():
  print(key, ":", value)
```

That last observation is sharp. You noticed something important without being prompted.

print(trade) dumps everything in one ugly line with curly brackets. The loop gives you one clean line per field — much more readable. In your ML notebooks, this exact pattern is used to print model scores, hyperparameters, and evaluation results in a clean format.

Your explanation of key and value is also exactly right. .items() unpacks each entry in the dictionary into two parts — the name and the value — and the loop handles one pair at a time.

Session 3 wrap-up
Dictionaries — store values with named keys instead of index numbers. Use {} and :

Access — dict["key"] instead of list[0]

Modify — dict["key"] = new_value

Add — same syntax, just use a new key name

Remove — del dict["key"]

Loop — .items() gives you key-value pairs one at a time

The one thing to remember:

Lists are ordered by position. Dictionaries are organised by name. Use a list when order matters, a dictionary when labels matter.

```python
result = {
    "instrument": "ES",
    "entry_price": 7394.25,
    "exit_price" : 7363.75,
    "contracts": 5,
    "direction": "short"
}

pnl = result["contracts"] * ((result["entry_price"] - result["exit_price"]))*4 * 12.5
print(pnl)
```

```python
result["P&L"] = pnl
print(result)
```

```python
for column, value in result.items():#.items and colon
  print(column, ":", value)
```

```python
trade1 = {
    "Direction": "Long",
    "Entry": 5842.25,
    "Exit": 5851.50,
    "Contracts": 3
}

trade2 = {
    "Direction": "Short",
    "Entry": 5856.00,
    "Exit": 5847.25,
    "Contracts": 2
}

trade3 = {
    "Direction": "Long",
    "Entry": 5848.75,
    "Exit": 5838.00,
    "Contracts": 4
}

```

```python
pnl = trade1["Contracts"] * ((trade1["Exit"] - trade1["Entry"])) * 50
trade1["Pnl"] = pnl
pnl = trade2["Contracts"] * ((trade2["Entry"] - trade2["Exit"])) * 50
trade2["Pnl"] = pnl
pnl = trade3["Contracts"] * ((trade3["Exit"] - trade3["Entry"])) * 50
trade3["Pnl"] = pnl
```

```python
journal = [trade1, trade2, trade3]
```

```python
print(journal[0]["Pnl"])
print(journal[1]["Pnl"])
print(journal[2]["Pnl"])
```

```python
total_pnl = journal[0]["Pnl"] + journal[1]["Pnl"] + journal[2]["Pnl"]
print("The total Profit and Loss is", total_pnl)
```
